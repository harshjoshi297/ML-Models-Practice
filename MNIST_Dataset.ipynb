```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential,layers
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
X_train = X_train/255
X_test = X_test/255 # Converting all values between 0 and 1.
```

```python
X_train[0]
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28))) #Flatten converts higher dim data to a 1-d array.
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
#Sparse-categorical cross entropy - we use it when we dont need to label encode the target variables.
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_split=0.1)
```

```python
y_prob = model.predict(X_test)
```

```python
y_prob
```

```python
y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
y_pred = y_prob.argmax(axis=1)
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'],color='red',label='train')
plt.plot(history.history['val_loss'],color='blue',label='validation')
plt.legend()
plt.show()
```
